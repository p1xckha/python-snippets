# display a flowchart using graphviz

https://graphviz.org/documentation/

## digraph

```python
# display a Markov Chain

import graphviz

dot_data = '''
digraph {
    rankdir=LR;  // Set the rank direction to Left to Right (LR)
    label="Markov Chain";  // Set the title of the graph
    0 [label="0"];
    1 [label="1"];
    2 [label="2"];
    // 3, 4 are recurrent states
    3 [label="3 - recurrent"];
    4 [label="4 - recurrent"];
    
    0 -> 0;
    0 -> 1;
    1 -> 2;
    1 -> 1;
    2 -> 3;
    2 -> 2;
    3 -> 3;
    3 -> 4;  
    4 -> 4;
    4 -> 3
}
'''
graph = graphviz.Source(dot_data)
graph
```

```python
import graphviz

dot_data = '''
    digraph {
    node [style="filled", shape="circle", width="0.5", fixedsize="true"];
    rankdir=LR;
    A [fillcolor="blue", label="A"];
    B [fillcolor="green", label="B", shape="cylinder"];
    C [fillcolor="orange", label="C"];
    D [fillcolor="purple", label="D"];
    E [fillcolor="red", label="E"];
    
    
    A -> {B C};
    B -> {C D};
    C -> {D E};
    D -> E;
    E -> E
}
'''
graph = graphviz.Source(dot_data)
graph
```

```python
dot_data = '''
digraph {
  node [shape=box];
  start -> A;
  A -> {B C}
  B -> C;
  C -> A;
  
}
'''

graph = graphviz.Source(dot_data)
graph
```

## graph

```python
dot_data = '''
graph {
    rankdir=LR;
    A -- B;
    B -- C;
    C -- A;
}
'''

graph = graphviz.Source(dot_data)
graph
```

# subgraph

```python
dot_data = '''
digraph {
    // Statements for the main graph
    A -> B;
    B -> C;
    
    subgraph cluster1 {
        // Statements for the subgraph
        style=filled;
        color=lightgray;
        node [shape=box];
        D -> E;
        E -> F;
    }
}
'''


graph = graphviz.Source(dot_data)
graph
```
