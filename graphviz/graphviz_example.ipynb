```python
import graphviz

dot_data = '''
digraph {
    rankdir=LR;  // Set the rank direction to Left to Right (LR)
    label="Markov Chain";  // Set the title of the graph
    0 [label="0"];
    1 [label="1"];
    2 [label="2"];
    3 [label="3"];
    4 [label="4 - recurrent"];
    0 -> 1;
    0 -> 0;
    1 -> 2;
    1 -> 1;
    2 -> 3;
    2 -> 2;
    3 -> 3;
    3 -> 4;  // 4 is a transient state
    4 -> 4
}
'''
graph = graphviz.Source(dot_data)
graph
```

```python
import graphviz

dot_data = '''
    digraph {
    rankdir=LR;
    node [style="filled", shape="circle", width="0.5", fixedsize="true"];
    
    A [fillcolor="blue", label="A"];
    B [fillcolor="green", label="B"];
    C [fillcolor="orange", label="C"];
    D [fillcolor="purple", label="D"];
    E [fillcolor="red", label="E"];
    
    A -> B;
    A -> C;
    B -> C;
    B -> D;
    C -> D;
    C -> E;
    D -> E;
}
'''
graph = graphviz.Source(dot_data)
graph
```
