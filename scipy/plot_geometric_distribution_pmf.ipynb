# Geometric distribution

- $K \sim \text{Geometric}(p)$
- $p: \text{parameter of the geometric distribution}; 0< p \leq 1$
- PMF of K  $P_K (k) = (1-p)^{k-1} p , (k=1, 2,\cdots)$
- $E[K] = \frac{1}{p}$
- $Var(K) = \frac{1-p}{p^2}$

https://en.wikipedia.org/wiki/Geometric_distribution


# Plot the Probability Mass Function of K

```python
import matplotlib.pyplot as plt
from scipy.stats import geom

# Parameter
p = 0.3  # Probability of success

# Calculate probabilities for each k
k_values = range(1, 11)  # Plotting up to 10 trials
probabilities = [geom.pmf(k, p) for k in k_values]

# Creating the figure and axis objects
fig, ax = plt.subplots(figsize=(10,5))

# Plotting the PMF
ax.bar(k_values, probabilities)
ax.set_xlabel('Number of Trials (k)')
ax.set_ylabel('Probability')
ax.set_title('Geometric Distribution PMF (p=0.3)')

# Calculate the mean
mean = 1 / p
variance = (1 - p) / (p**2)
std_deviation = variance**0.5

# Display the Mean, Variance, Standard deviation
ax.text(0.7, 0.5, f'Mean = {mean:.2f}', transform=ax.transAxes)
ax.text(0.7, 0.4, f'Variance = {variance:.2f}', transform=ax.transAxes)
ax.text(0.7, 0.3, f'Standard deviation = {std_deviation:.2f}', transform=ax.transAxes)

# Draw vertical line at k = mean, mean-sigma, mean+sigma
ax.axvline(x=mean, color='r', linestyle='--', label='Mean')
ax.axvline(x=mean + std_deviation, color='b', linestyle=':', label='Mean + Standard Deviation')
ax.axvline(x=mean - std_deviation, color='g', linestyle=':', label='Mean - Standard Deviation')

# Display the legend
ax.legend()

# Display the plot
plt.show()
```
