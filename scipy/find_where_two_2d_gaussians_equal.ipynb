# Find $x \in \mathbf{R}^2$ where $f_{\mu_1, \Sigma_1} (x) = f_{\mu_2, \Sigma_2} (x)$
* If $\Sigma_1 = \Sigma_2 = \Sigma $: a straight line  
* If $\Sigma_1 \neq \Sigma_2$ : a quadratic form 

# If $\Sigma_1 \neq \Sigma_2$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Parameters for the first Gaussian distribution
mu1 = np.array([1, 1])
S1 = np.array([[2, 1], 
               [1, 2]])

# Parameters for the second Gaussian distribution
mu2 = np.array([2, 2])
S2 = np.array([[1, 1], 
               [1, 4]])

def plot(mu1, mu2, S1, S2, xmin=-1, xmax=5, ymin=-1, ymax=5):
    # Create a grid of values for the x and y coordinates
    x, y = np.meshgrid(np.linspace(xmin, xmax, 400), np.linspace(ymin, ymax, 400))
    pos = np.dstack((x, y))
    
    # Calculate the probability densities for both distributions at each point on the grid
    pdf1 = multivariate_normal.pdf(pos, mean=mu1, cov=S1)
    pdf2 = multivariate_normal.pdf(pos, mean=mu2, cov=S2)
    
    # Create contour plots for both distributions
    fig, ax = plt.subplots(figsize=(8, 8))
    contour1 = ax.contour(x, y, pdf1, colors='b', levels=5)  # Blue contours for the first distribution
    contour2 = ax.contour(x, y, pdf2, colors='r', levels=5)  # Red contours for the second distribution
    
    # Calculate the differences between the two distributions
    differences = pdf1 - pdf2
    contour = ax.contour(x, y, differences, levels=[0], colors='c')  
    
    # Scatter plot for mu1 and mu2 with legends
    ax.scatter(mu1[0], mu1[1], label='$\mu_1$', color='b', marker='o')
    ax.scatter(mu2[0], mu2[1], label='$\mu_2$', color='r', marker='o')
    
    # Create custom handles for the legend
    legend1 = plt.Line2D([0], [0], color='b', lw=2, label='$N(\mu_1, \Sigma_1)$')
    legend2 = plt.Line2D([0], [0], color='r', lw=2, label='$N(\mu_2, \Sigma_2)$')
    legend3 = plt.Line2D([0], [0], color='c', label='Where two PDFs equal')
    legend4 = plt.Line2D([0], [0], color='b', marker='o', markersize=5, label='$\mu_1$', linestyle='None')
    legend5 = plt.Line2D([0], [0], color='r', marker='o', markersize=5, label='$\mu_2$', linestyle='None')
    
    # Add the custom legend to the plot
    ax.legend(handles=[legend1, legend2, legend3, legend4, legend5])
    
    # You can also add labels and a title if needed
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_title('Contour Plot of Multivariate Gaussian Distributions')
    
    ax.grid(True)
    plt.show()

plot(mu1, mu2, S1, S2)
```

# If $\Sigma_1 = \Sigma_2 = \Sigma $

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Parameters for the first Gaussian distribution
mu1 = np.array([1, 1])

# Parameters for the second Gaussian distribution
mu2 = np.array([2, 1])

S = np.array([[1, -1], 
               [-1, 4]])

plot(mu1, mu2, S, S, ymin=-3, ymax=6)
```
