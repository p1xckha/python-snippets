#  compute cumulative probability of Normal random variable X

$$X \sim N(80, 25)$$
Find $P(X \leq 82)$

# answer
$$
\begin{eqnarray*}
P(X \leq 82) 
& = & P(Z \leq (82-80)/5) \\
& = & P(Z \leq 0.4) \\
& = & 0.65542
\end{eqnarray*}
$$
https://en.wikipedia.org/wiki/Standard_normal_table


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the range for the x-axis (Z-values)
x = np.linspace(-4, 4, 1000)

# parameters for normal  
mu = 80
sigma = 5

# Calculate the standard normal distribution values (PDF) for the given x
standard_normal_pdf = stats.norm.pdf(x, 0, 1)

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10,6))

# Plot the standard normal distribution on ax
ax.plot(x, standard_normal_pdf, label='PDF of $Z$ $\sim$ $N(0,1)$')

# Highlight the area corresponding to P(Z <= z_score)
z_score = (82 - mu)/sigma
x_shaded = np.linspace(-4, z_score, 100)
y_shaded = stats.norm.pdf(x_shaded, 0, 1)
ax.fill_between(x_shaded, y_shaded, color='blue', alpha=0.3)

# Draw the vertical line at Z = z_score
ax.axvline(z_score, color='red', linestyle='--', label='Z = 1.622')

# Set plot labels and title
ax.set_xlabel('Z')
ax.set_ylabel('Probability Density')
ax.set_title(f'PDF of the Standard Normal with shaded region $Z \leq {z_score}$')
ax.legend()

# Show the plot
ax.grid(True)
plt.show()

# Calculate the cumulative probability P(Z <= z_score) using the cumulative distribution function (CDF)
cumulative = stats.norm.cdf(z_score)

print(f"P(Z <= {z_score}) = {cumulative:.5f}")
```
