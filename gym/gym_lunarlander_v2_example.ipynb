```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

env = gym.make("LunarLander-v2", render_mode="rgb_array")
env.reset()
terminated = False
total_reward = 0.0
env_screens = []
while not terminated:
    # choose a ramdom action
    action = env.action_space.sample() 
    # tak action
    observation, reward, terminated, truncated, info = env.step(action) 
    # prepare next step
    total_reward += reward
    env_screen = env.render() # numpy.ndarray
    env_screens.append(env_screen)

env.close()
total_reward
```

```python
fig, ax = plt.subplots()
im = ax.imshow(env_screens[0])

def update(i):
    im.set_data(env_screens[i])
    return (im)

animation = FuncAnimation(fig=fig, func=update, frames=len(env_screens), interval=100)
#plt.show()

# display the animation in the notebook
HTML(animation.to_jshtml())
```

```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML


parameters = {
    'continuous': True,
    'gravity': -11.99,
    'enable_wind': True,
    'wind_power': 0.0001,
    'turbulence_power': 0.0000001,
}

ENV_NAME = "LunarLander-v2"
env = gym.make(ENV_NAME, **parameters, render_mode="rgb_array")

env.reset()
terminated = False
total_reward = 0.0
env_screens = []
while not terminated:
    action = np.array([0.1 , -0.9]) # main and left booster fire
    # tak action
    observation, reward, terminated, truncated, info = env.step(action) 
    # prepare next step
    total_reward += reward
    env_screen = env.render() # numpy.ndarray
    env_screens.append(env_screen)

env.close()
total_reward
```

```python
####################################################
# create an animation
#===================================================

fig, ax = plt.subplots()
im = ax.imshow(env_screens[0])

def update(i):
    im.set_data(env_screens[i])
    return (im)

animation = FuncAnimation(fig=fig, func=update, frames=len(env_screens), interval=100)
#plt.show()

# display the animation in the notebook
HTML(animation.to_jshtml())
```
