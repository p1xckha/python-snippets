```python
########################################
# display the animation in the notebook
########################################
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# two points moves on the circle
fig, ax = plt.subplots()
theta = np.arange(-2*np.pi, 2*np.pi, 0.1)
r = 1.0
x, y = r*np.cos(theta), r*np.sin(theta) 
fast_x, fast_y = r*np.cos(2*theta), r*np.sin(2*theta) 

slow_point = ax.scatter(x[0], y[0], c="b", s=80)
fast_point = ax.scatter(fast_x[0], fast_y[0], c="g", s=10)

line2 = ax.plot(x, y, c="r", linewidth=0.25)
ax.axis('off') # turn off the axis visibility
ax.set(xlim=[-1.1, 1.1], ylim=[-1.1, 1.1])

def update(frame):
    # update the scatter plot
    slow_point.set_offsets(np.c_[x[frame], y[frame]])
    fast_point.set_offsets(np.c_[fast_x[frame], fast_y[frame]])
    return (slow_point, fast_point, line2)

animation = FuncAnimation(fig=fig, func=update, frames=len(x), interval=100)
plt.show()

# display the animation in the notebook
HTML(animation.to_jshtml())
```
