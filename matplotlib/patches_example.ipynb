```python
##################################################
# Rectangle
#=================================================

# https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Rectangle.html#matplotlib.patches.Rectangle

import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots()

# create a rectangle patch
xy = (3, 3) # usually the bottom left corner
width = 5
height = 4
kwargs = {'edgecolor':'g', 'facecolor':'none'} # optional
rectangle = patches.Rectangle(xy, width, height, linewidth=2, **kwargs)

# add the patch to the axes
ax.add_patch(rectangle)
ax.axis('equal')

plt.show()
```

```python
##################################################
# Circle
#=================================================

# https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Circle.html#matplotlib.patches.Circle

import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots()

# create a circle patch
xy = (0, 0) 
radius = 10 # default 5
kwargs = {'edgecolor':'r', 'facecolor':'r'} # optional

circle = patches.Circle(xy, radius=radius, **kwargs)

# add the patch to the axes
ax.add_patch(circle)
ax.axis('equal')

plt.show()
```

```python
##################################################
# Annulus
#=================================================

# https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Annulus.html#matplotlib.patches.Annulus

import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots()

xy = (0, 0) 
r = 2.0
width = 1.0
annulus = patches.Annulus(xy, r, width, edgecolor='brown', facecolor='brown')

# add the patch to the axes
ax.add_patch(annulus)
ax.axis('equal')

plt.show()
```

```python
##################################################
# Arrow
#=================================================
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Arrow.html#matplotlib.patches.Arrow

fig, ax = plt.subplots()

x, y = 0.0, 0.0
dx, dy = 3, 4
arrow = patches.Arrow(x, y, dx, dy)

ax.add_patch(arrow)
ax.set(xlim=[-1,5], ylim=[-1, 5])
plt.show()
```

```python
##################################################
# FancyArrow
#=================================================
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.FancyArrow.html#matplotlib.patches.FancyArrow

fig, ax = plt.subplots()

x, y = 0.0, 0.0
dx, dy = 3, 4
width=0.1

# you can set head width and head height independently
head_width = 0.3
head_length = 0.3
fancy_arrow = patches.FancyArrow(x, y, dx, dy, width=width, head_width=head_width, head_length=head_length)

ax.add_patch(fancy_arrow)
ax.set(xlim=[-1,5], ylim=[-1, 5])
plt.show()
```

```python
##################################################
# Polygon
#=================================================

# https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Polygon.html#matplotlib.patches.Polygon

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

fig, ax = plt.subplots()
n = 3
xy = np.array([[0.0, 0.0], [0.0, 1.0], [1.0,1.0], [2.0, 0.0]])
polygon = patches.Polygon(xy)

# add the patch to the axes
ax.add_patch(polygon)
ax.axis('equal')

plt.show()

```

```python
##################################################
# CirclePolygon
#=================================================

# https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.CirclePolygon.html#matplotlib.patches.CirclePolygon

import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots()

xy = (0, 0) 
radius=10
resolution=8
circle_polygon = patches.CirclePolygon(xy, radius=radius, resolution=resolution)

# add the patch to the axes
ax.add_patch(circle_polygon)
ax.axis('equal')

plt.show()
```

```python
##################################################
# RegularPolygon
#=================================================

# https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.RegularPolygon.html#matplotlib.patches.RegularPolygon

import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots()

xy = (5, 5)  # The center position
radius = 3
numVertices = 5 
regular_polygon = patches.RegularPolygon(xy, numVertices, radius=5)

# add the patch to the axes
ax.add_patch(regular_polygon)
ax.axis('equal')

plt.show()


```
