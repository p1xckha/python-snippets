# relationship between gradient and normal vector

## 3D Plane
$$z = f(x,y)= ax+by$$

## normal vector
$$n=(a,b,-1)$$
[Normal (geometry) - Wikipedia](https://en.wikipedia.org/wiki/Normal_(geometry))

## gradient f
$$\nabla_{x,y} f(x,y) = (a, b)$$
grad(f) is the steepest direction on the plane. I explain why.

$$\Delta x = \cos \theta , \Delta y = \sin \theta $$
then, 
$$\Delta z(\theta) =  a \cos \theta + b \sin \theta $$
taking derivative of delta z w.r.t. theta.
$$ \frac{\partial \Delta z(\theta)}{\partial \theta} = -a\sin \theta + b \cos \theta = 0$$
$$ \Delta x : \Delta y = cos\theta : sin \theta = a:b$$



# Illustrate 3D Plane, Gradient, normal vector

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create the figure and 3D axis
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')

# Generate x, y values for the plane
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
a, b, c = 1, 2, 0
# Calculate z values based on the plane equation
Z = a * X + b * Y + c

# Plot the 3D-plane
ax.plot_surface(X, Y, Z, alpha=0.5)

# Draw the vector (1, 2, 0) from the origin on the xy-plane with a bigger head
ax.quiver(0, 0, 0, a, b, 0, color='red', length=1.5, arrow_length_ratio=0.4, linewidth=2, label='grad(f)@(0,0)=(1,2)')

# Draw the vector (1, 2, -1) from the origin on the xy-plane with a bigger head
ax.quiver(0, 0, 0, a, b, -1, color='blue', length=1.5, arrow_length_ratio=0.4, linewidth=2, label='normal vector from (0,0,0)=(1,2,-1)')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plane: f(x,y) = x + 2y')
ax.view_init(elev=10, azim=-15)

# Set axis limits
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-5, 5)

ax.axis('equal')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
