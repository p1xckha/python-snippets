# Triangle inequality

$$||u| - |v|| \leq |u-v| \leq |u| + |v|$$
where $u,v \in \mathbb{C}$ 

https://en.wikipedia.org/wiki/Triangle_inequality
    

```python
import numpy as np
import matplotlib.pyplot as plt

# Complex numbers u1, v1, and u2
u1 = 3 * np.exp(1j * np.pi / 4)
v1 = 4 * np.exp(1j * 3 * np.pi / 4)
u2 = 3 * np.exp(-1j * np.pi / 4)

# Additional points u3 and v2
u3 = 3
v2 = 4
v3 = 4 * np.exp(1j * 7 * np.pi /6)

# Generate a set of points along the first circle
theta1 = np.linspace(0, 2 * np.pi, 100)
circle1 = 3 * np.exp(1j * theta1)

# Generate a set of points along the second circle
theta2 = np.linspace(0, 2 * np.pi, 100)
circle2 = 4 * np.exp(1j * theta2)

# Separate the real and imaginary parts for plotting
u1_real = np.real(u1)
u1_imag = np.imag(u1)

v1_real = np.real(v1)
v1_imag = np.imag(v1)

u2_real = np.real(u2)
u2_imag = np.imag(u2)

v3_real = np.real(v3)
v3_imag = np.imag(v3)

u3_real = u3
u3_imag = 0

v2_real = v2
v2_imag = 0

circle1_real = np.real(circle1)
circle1_imag = np.imag(circle1)

circle2_real = np.real(circle2)
circle2_imag = np.imag(circle2)

# Create a larger figure
fig, ax = plt.subplots(figsize=(12, 12))

# Plot the circles and points
ax.plot(circle1_real, circle1_imag, label='Circle (r=3)')
ax.plot(circle2_real, circle2_imag, label='Circle (r=4)')
ax.plot(u1_real, u1_imag, 'ro', label='u1 (3*e^(j*pi/4))')
ax.plot(v1_real, v1_imag, 'bo', label='v1 (4*e^(j*3*pi/4))')
ax.plot(u2_real, u2_imag, 'go', label='u2 (3*e^(-j*pi/4))')

# Scatter plot for points u3 and v2
ax.scatter(u3_real, u3_imag, color='orange', label='u3')
ax.scatter(v2_real, v2_imag, color='purple', label='v2')
ax.scatter(v3_real, v3_imag, color='c', label='v3')

# Draw lines from origin to u1, v1, u2, u3, and v2 using solid lines
ax.plot([0, u1_real], [0, u1_imag], 'r', linestyle='solid')
ax.plot([0, v1_real], [0, v1_imag], 'b', linestyle='solid')
ax.plot([0, u2_real], [0, u2_imag], 'g', linestyle='solid')
ax.plot([0, v3_real], [0, v3_imag], 'c', linestyle='solid')


ax.plot([u3_real, v2_real], [u3_imag, v2_imag], 'g', linestyle='solid')

# Add line and text for |u2-v1|
u2_v1 = u2 - v1
ax.plot([u2_real, v1_real], [u2_imag, v1_imag], 'purple', linestyle='solid')
ax.text(-1.5 , 0.5 , '$|u|+|v|$', color='purple', fontsize=16)
ax.text(3.2 , -0.25 , '$||u|-|v||$', color='g', fontsize=16)

ax.text(-2, -1, '$|v|$', color='c', fontsize=16)
ax.text(1.25, 1, '$|u|$', color='r', fontsize=16)



# Add texts near the points u1, v1, u2, u3, and v2
offset = 0.2
ax.text(u1_real + offset, u1_imag + offset, 'u1', color='r', fontsize=16)
ax.text(v1_real + offset, v1_imag + offset, 'v1', color='b', fontsize=16)
ax.text(u2_real + offset, u2_imag + offset, 'u2', color='g', fontsize=16)
ax.text(u3_real + offset, u3_imag + offset, 'u3', color='orange', fontsize=16)
ax.text(v2_real + offset, v2_imag + offset, 'v2', color='purple', fontsize=16)
ax.text(v3_real + offset, v3_imag + offset, 'v3', color='c', fontsize=16)

# Set axis limits to include all points and circles
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

# Add labels and legend
ax.set_xlabel('Real part', fontsize=16)
ax.set_ylabel('Imaginary part', fontsize=16)
ax.set_title('triangle inequality $||u| - |v|| \leq |u-v| \leq |u| + |v|$', fontsize=16)
ax.legend()
ax.grid(True)
ax.axis('equal')  # Set the aspect ratio to be equal

# Show the plot
plt.show()
```
