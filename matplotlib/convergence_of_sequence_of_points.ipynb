```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f(x):
    return 1-0.25*np.exp(x)

def g(x):
    return x

# Generate x values for the plot
x = np.linspace(0, 1.5, 500)

# Calculate y values for both functions
y1 = f(x)
y2 = g(x)

# Generate fig and ax
fig, ax = plt.subplots(figsize=(10,10))

# Plot the functions
ax.plot(x, y1, label='$y=1- \\frac{1}{4}\exp(x_n)$')
ax.plot(x, y2, label='$y = x$')

# Plot the lines
x0 = 0 # Choose a value for x0
for i in range(5):
    x1 = f(x0)
    # Arrow from (x0,x0) to (x0, x1)
    arrow1 = plt.Arrow(x0, x0, x0 - x0, x1 - x0, width=0.03, color='c')
    # Arrow from (x0,x1) to (x1, x1)
    arrow2 = plt.Arrow(x0, x1, x1 - x0, x1 - x1, width=0.03, color='c')
    # Add arrows to the graph
    ax.add_patch(arrow1)
    ax.add_patch(arrow2)
    # Prepare next step
    x0 = x1

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Illustration of the Convergence of $x_{n+1} = 1 - \\frac{1}{4} \exp(x_n)$')

# Add a legend to differentiate the functions and lines
ax.legend()

# Show the plot
ax.axis('equal')
ax.grid(True)

plt.show()

print("i={}".format(x1)) # i=3.809589230764377
```
