# plot grad(f)

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

##########################
# f(x, y) = x**2 + y**2
##########################

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z)

##########################
# grad(f) = (2*x, 2*y)
##########################
x = np.linspace(-2, 2, 5)
y = np.linspace(-2, 2, 5)
z = np.linspace(0, 0, 1) # array([0])
X, Y, Z = np.meshgrid(x, y, z)

# gradient
grad_x = 2*X
grad_y = 2*Y
w = np.zeros_like(X)

# draw gradients on the xy plane
ax.quiver(X, Y, Z, grad_x, grad_y, w, length=0.2, color='green')

##
ax.axis('equal')
ax.set(xlim=[-3, 3], ylim=[-3, 3])
ax.set_title('f(x,y)=x**2 + y**2, grad(f)=(2*x, 2*y)', fontsize=20)
```

# plot normal vectors 

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

##########################
# f(x, y) = -(x**2 + y**2)
##########################

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)


X, Y = np.meshgrid(x, y)
Z = -(X**2 + Y**2)
ax.plot_surface(X, Y, Z)


##########################
# normal vector = -(-2*x, -2*y, -1) 
##########################

x = np.linspace(-2, 2, 4)
y = np.linspace(-2, 2, 4)
X, Y = np.meshgrid(x, y)

# draw normal vectors on the plane
Z = -(X**2 + Y**2)

# upward normal vectors
normal_x = 2*X
normal_y = 2*Y
w = np.ones_like(X) 
ax.quiver(X, Y, Z, normal_x, normal_y, w, length=0.4, color='red')

##
ax.axis('equal')
ax.set_title('f(x,y)= -(x**2 +y**2), normal vector=(2*x, 2*y, 1)', fontsize=20)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the viewing angle
ax.view_init(elev=10, azim=-100)
```
