# quiver  example
[matplotlib.axes.Axes.quiver â€” Matplotlib 3.7.1 documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.quiver.html#matplotlib.axes.Axes.quiver)

```python
import numpy as np
import matplotlib.pyplot as plt

##########################
# f(x, y) = x**2 + y**2
##########################

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)

X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)

##########################
# grad(f) = (2*x, 2*y)
##########################
x = np.linspace(-1, 1, 5)
y = np.linspace(-1, 1, 5)
z = np.linspace(0, 0, 1) # array([0])
X, Y, Z = np.meshgrid(x, y, z)

# gradient
grad_x = 2*X
grad_y = 2*Y
w = np.zeros_like(X)

# draw gradients on the xy plane
ax.quiver(X, Y, Z, grad_x, grad_y, w, length=0.2, color='green')

ax.set(xlim=[-1.5, 1.5], ylim=[-1.5, 1.5])
ax.set_title('f(x,y)=x**2 + y**2, grad(f)=(2*x, 2*y)', fontsize=20)
```
