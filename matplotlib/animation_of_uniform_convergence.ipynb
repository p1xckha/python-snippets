# uniform convergence example

## sequence of functions Fn
$$F_n(x) = 1 - (1 - 1/n)^{0.5 n x}$$

## limit function F(x)
$$F(x)=1-e^{-0.5x} $$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display

# Cumulative Density Function
def Fxn(x, n):
    if x < 0:
        return 0
    else:
        return 1 - (1 - 1/n)**(0.5*n*x)

def limit_func(x):
    if x < 0:
        return 0
    else:
        return 1 - np.exp(-0.5*x)    

def update(frame):
    ax.cla()  # Clear the current plot

    n = frame + 2  # Start with n=2 and increment by 1 for each frame
    x = np.linspace(-0.5, 10, 1000)
    y = np.array([Fxn(val, n) for val in x])
    y0 = np.array([limit_func(val) for val in x])

    ax.plot(x, y, label=f'F_{n}(x) = 1 - (1 - 1/{n})^({n}x)')
    ax.plot(x, y0, label='limit function $F(x)=1-e^{-0.5x}$')
    ax.set_xlabel('x')
    ax.set_ylabel(f'F_{n}(x)')
    ax.set_title(f'F_{n}(x) = 1 - (1 - 1/{n})^({n}x)', fontsize=16)
    ax.grid(True)
    ax.legend()
    ax.axhline(0, color='black', linewidth=0.5, linestyle='dashed')  # x-axis
    ax.axvline(0, color='black', linewidth=0.5, linestyle='dashed')  # y-axis
    
    # Fix the range of the x-axis and y-axis for consistency
    ax.set_xlim(-0.5, 10)
    ax.set_ylim(-0.1, 1.1)

# Set up the figure and animation
fig, ax = plt.subplots(figsize=(10, 7))
ani = FuncAnimation(fig, update, frames=30, interval=250)

# Convert the animation to HTML format and display it in the notebook
HTML(ani.to_jshtml())
```
