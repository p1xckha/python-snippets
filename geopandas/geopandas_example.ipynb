```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib
import geodatasets
from geodatasets import get_path

# matplotlib setting
matplotlib.rcParams["svg.fonttype"] = "none"
matplotlib.rcParams["font.family"] = "sans-serif"
matplotlib.use("SVG")
```

```python
# dataset names
geodatasets.data.flatten().keys()
```

# The map of the world

```python
# draw the map of the world
dataset_name = "naturalearth.land"
world = gpd.read_file(get_path(dataset_name))
fig, ax = plt.subplots()
world.plot(color="white", edgecolor="black", ax=ax)

# title
title_url="https://en.wikipedia.org/wiki/World_map"
ax.set_title('The Map of The World', url=title_url)

# labels
longitude_url = "https://en.wikipedia.org/wiki/Longitude"
latitude_url = "https://en.wikipedia.org/wiki/Latitude"
ax.set_xlabel('Longitude', url=longitude_url)
ax.set_ylabel('Latitude', url=latitude_url)
ax.grid(True, linestyle='--', alpha=0.7)
fig
```

# Save the world map in svg format

```python
svg_filename = "world.svg"
fig.savefig(svg_filename)

import os
svg_path = os.path.join(os.getcwd(), svg_filename)
file_info = os.stat(svg_path)
print(svg_path, str(file_info.st_size/1000) + "KB")
```

# Clip the world map

```python
lon_min, lat_min, lon_max, lat_max = -100, -60, -30, 25
south_america = world.clip([lon_min, lat_min, lon_max, lat_max])
fig, ax = plt.subplots()
south_america.plot(ax=ax, color='lightgreen', edgecolor='green')
ax.set_title('South America')
fig
```

```python
newzealand = world.clip([160, -50, 180, -25])
fig, ax = plt.subplots()
newzealand.plot(ax=ax, color='lightgreen', edgecolor='green')
ax.set_title('New Zealand')
fig
```
