```python
world = gpd.read_file(get_path("naturalearth.land"))
ax = world.plot(color="white", edgecolor="black")
ax.set_title('The Map of The World')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
lon_min, lat_min, lon_max, lat_max = -100, -60, -30, 25
south_america = world.clip([lon_min, lat_min, lon_max, lat_max])
ax = south_america.plot(color='lightgreen', edgecolor='green')
```

```python
newzealand = world.clip([160, -50, 180, -25])
ax = newzealand.plot(color='lightgreen', edgecolor='green')
```
