# Circulant graph

* https://en.wikipedia.org/wiki/Circulant_graph

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np

# Construct an adjacency matrix
n = 12
A = np.zeros([n,n])
for i in range(n):
    for j in range(n):
        # connect four neigbors
        if abs(i-j) % n == 1 or abs(i-j) % n == 2 or abs(i-j) % n == n-1 or abs(i-j) % n == n-2:
            A[i,j] = 1

# Create a graph
G = nx.Graph(A)

# Create a figure for the plot
fig, ax = plt.subplots(figsize=(10,10))

# Create a circular layout for the nodes
pos = nx.circular_layout(G)

# Draw the graph in a circular layout
nx.draw(G, pos, with_labels=True, node_size=400, node_color='skyblue', ax=ax, width=0.5)

# Step 5: Set the title and labels (optional)
ax.set_title("Circulant Graph", size=20)

# Display the plot
plt.show()
```

# Do the same thing with nx.circulant_graph()
https://networkx.org/documentation/stable/reference/generated/networkx.generators.classic.circulant_graph.html

```python
n = 10
# connect i and i+1, i and i-1, i and i+2, i and i-2
G = nx.circulant_graph(n, [1,2])

fig, ax = plt.subplots(figsize=(10,10))
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_size=400, node_color='skyblue', ax=ax, width=0.5)
ax.set_title("Circulant Graph", size=20)
plt.show()
```
